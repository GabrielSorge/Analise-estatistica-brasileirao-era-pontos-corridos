```python
df["#DATE"] = pd.to_datetime(df["#DATE"])
df["#DATE"]
```

```python

def verifica_resultado(resultado):
  placar = resultado.split("x")

  casa = int(placar[0])
  visitante = int(placar[1])

  if casa > visitante:
    resultado_visitante = "derrota"
  elif casa == visitante:
    resultado_visitante = "empate"
  else:
    resultado_visitante = "vitoria"

  return resultado_visitante
```

```python
df["RESULTADO_VISITANTE"] = df["SCORE"].apply(verifica_resultado)
df["RESULTADO_VISITANTE"]
```
